### For Page 2-258 ( LOOP IMPLEMENTATION)

```python
for i in range(1,259):
    x = tables[i].df[2:].copy()
    x = x.replace('\n','', regex=True)
    l = len(x)
    for j in range(2,l):
        if(x.at[j,4] == "Aziz Bhatti Guldasht Town"):
            x.at[j,3] = "Aziz Bhatti"
            x.at[j,4] = "Guldasht Town"
    tables[i].df = x


    
```

### For Page 259 and onwards

```python
for i in range(259,len(tables)):
    x = tables[i].df.copy()
    x = x.replace('\n','', regex=True)
    l = len(x)
    for j in range(l):
        if(x.at[j,4] == "Aziz Bhatti Guldasht Town"):
            x.at[j,3] = "Aziz Bhatti"
            x.at[j,4] = "Guldasht Town"
    tables[i].df = x
```
